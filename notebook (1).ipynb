## Test statistical hypotheses

[Test the hypothesis that the average revenue from users of the Ultimate and Surf calling plans differs.]

N_0: The mean of plan Surf is the same as the mean of the plan Ultimate 
N_A: The mean of plan Surf is not the same as the mean of the plan Ultimate


[Formulate the null and the alternative hypotheses, choose the statistical test, decide on the alpha value.]

````python
# Test the hypotheses

from scipy import stats as st
surf_revenue = merged_df_2[merged_df_2['plan_name'] == 'surf']['revenue']
ultimate_revenue = merged_df_2[merged_df_2['plan_name'] == 'ultimate']['revenue']
# print(surf_revenue)
# print(ultimate_revenue)
alpha = .05#
results = st.ttest_ind(surf_revenue, ultimate_revenue, equal_var=False)
print(results)
print('p-value: ', results.pvalue)

if results.pvalue < alpha:
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis")
````
